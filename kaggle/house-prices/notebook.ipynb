Check for nulls:

```python
na_values = train_df.isna().sum().sort_values(ascending=False)
na_values = na_values[na_values > 0]

plt.figure(figsize=(12,6))
plt.bar(na_values.index, na_values.values)
plt.xticks(ticks=range(len(na_values)), labels=na_values.index, rotation=90)
plt.show()
```

```python
categorical_vars = [var for var in train_df.columns if train_df[var].dtype=='O'] 
cat = train_df.select_dtypes(include='object')
print(f"There are {len(categorical_vars)} categorical variables")
```

```python
numerical_vars = [var for var in train_df.columns if train_df[var].dtype!='O'] 
print(f"There are {len(numerical_vars)} numerical variables")
```

```python
obj_cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 
                'FireplaceQu', 'GarageType', 
                'GarageFinish', 'GarageQual', 'BsmtFinType2',
                'BsmtExposure', 'BsmtQual', 'BsmtFinType1',
                'GarageCond', 'BsmtCond'
                ]

train_df[obj_cols] = train_df[obj_cols].fillna("None")
```

```python
num_cols = ['GarageYrBlt', 'MasVnrArea']
train_df[num_cols] = train_df[num_cols].fillna(0)
```

```python
mode_fill = ['LotFrontage', 'MasVnrType', 'Electrical']
train_df[mode_fill] = train_df[mode_fill].fillna(train_df[mode_fill].mode())
```

```python
plt.figure(figsize=(16,11))
num_attributes = train_df.select_dtypes(include=['int64','float64'])
corr = num_attributes.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr, annot=True,
            mask=mask, cmap='mako',
            fmt='.1f', cbar=True)
```

```python
plt.figure(figsize=(16, 11))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True
           )
```
