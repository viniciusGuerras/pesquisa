```python
categorical_vars = [var for var in train_df.columns if train_df[var].dtype=='O'] 
print(f"There are {len(categorical_vars)} categorical variables")
```

```python
numerical_vars = [var for var in train_df.columns if train_df[var].dtype!='O'] 
print(f"There are {len(numerical_vars)} numerical variables")
```

```python
# Initial drop of problematic sparse columns
sparse_cols = ["PoolQC", "Fence", "MiscFeature", "Alley", "FireplaceQu"]
train_df = train_df.drop(columns=sparse_cols, errors="ignore")

# 1. Handle Numerical Columns: Fill NaNs with median
num_cols = train_df.select_dtypes(include=['int64', 'float64']).columns
train_df[num_cols] = train_df[num_cols].fillna(train_df[num_cols].median())

# 2. Handle Categorical Columns: Fill NaNs and Encode
cat_cols = train_df.select_dtypes(include=['object', 'category']).columns
for col in cat_cols:
    train_df[col] = train_df[col].fillna("Missing")
    train_df[col], _ = train_df[col].factorize()

# 3. Explicitly create the discrete_features mask
# All columns in train_df are now numeric (floats or integers from factorize)
# We treat object/category-converted-to-int columns as discrete
discrete_features = [col in cat_cols for col in train_df.columns]

# 4. Final check for any remaining NaNs (should be none)
if train_df.isnull().any().any():
    train_df = train_df.fillna(0)
```

```python

def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")
```

```python
def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features, random_state=0)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

X = train_df.copy()
y = np.log(X.pop("SalePrice"))
discrete_features = X.dtypes == "object"  
mi_scores = make_mi_scores(X, y, discrete_features)

# Plot
print(mi_scores.head(20))
plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(mi_scores.head(20))
plt.show()
```
